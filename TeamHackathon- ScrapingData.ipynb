```python
from selenium.webdriver.support.ui import Select

## Total points Scored

element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
TotalPoints=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in TotalPoints:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
MatchesPlayed=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-matchedPlayed'})
N=[]
for i in MatchesPlayed:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    N.append(k)
    
Season7_TeamPoints=pd.DataFrame(T,P)
Season7_TeamPoints=Season7_TeamPoints.reset_index()
Season7_TeamPoints=Season7_TeamPoints.rename(columns={'index':'Total Points',0        :'Team'})
Season7_TeamPoints=Season7_TeamPoints[['Team','Total Points']]
Season7_TeamPoints['MatchesPlayed']=N


## Total Points Conceded


element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('Total points conceded')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
TotalPointsConceded=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in TotalPointsConceded:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_TeamPointsConceded=pd.DataFrame(T,P)
Season7_TeamPointsConceded=Season7_TeamPointsConceded.reset_index()
Season7_TeamPointsConceded=Season7_TeamPointsConceded.rename(columns={'index':'Total Points Conceded',0        :'Team'})
Season7_TeamPointsConceded=Season7_TeamPointsConceded[['Team','Total Points Conceded']]

## Successful Raids


element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('Successful raids')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
successful_raids=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in successful_raids:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_successful_raids=pd.DataFrame(T,P)
Season7_successful_raids=Season7_successful_raids.reset_index()
Season7_successful_raids=Season7_successful_raids.rename(columns={'index':'Successful Raids',0        :'Team'})
Season7_successful_raids=Season7_successful_raids[['Team','Successful Raids']]


## Successful tackles


element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('Successful tackles')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
successful_tackles=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in successful_tackles:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_successful_tackles=pd.DataFrame(T,P)
Season7_successful_tackles=Season7_successful_tackles.reset_index()
Season7_successful_tackles=Season7_successful_tackles.rename(columns={'index':'Successful Tackles',0        :'Team'})
Season7_successful_tackles=Season7_successful_tackles[['Team','Successful Tackles']]

## SUPER Tackles


element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('Super tackles')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
super_tackles=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in super_tackles:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_super_tackles=pd.DataFrame(T,P)
Season7_super_tackles=Season7_super_tackles.reset_index()
Season7_super_tackles=Season7_super_tackles.rename(columns={'index':'Super Tackles',0        :'Team'})
Season7_super_tackles=Season7_super_tackles[['Team','Super Tackles']]


## SUPER RAID

element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('Super raid')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
super_raid=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in super_raid:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_super_raid=pd.DataFrame(T,P)
Season7_super_raid=Season7_super_raid.reset_index()
Season7_super_raid=Season7_super_raid.rename(columns={'index':'Super Raid',0        :'Team'})
Season7_super_raid=Season7_super_raid[['Team','Super Raid']]



## All-outs-inflicted

element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('All-outs inflicted')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
inflicted=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in inflicted:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_inflicted=pd.DataFrame(T,P)
Season7_inflicted=Season7_inflicted.reset_index()
Season7_inflicted=Season7_inflicted.rename(columns={'index':'Inflicted',0        :'Team'})
Season7_inflicted=Season7_inflicted[['Team','Inflicted']]

## All-outs-Conceded


element = Select (driver.find_element_by_id('si_dropdown'))
element.select_by_visible_text('All-outs conceded')
time.sleep(5)
element=driver.find_element_by_id('load_more')
element.click()
soup=BeautifulSoup(driver.page_source)
Team_names=soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
T=[]
for i in Team_names:
    h=i.text
    h=h.strip('\n')
    h=h.strip(' ')
    h=h.strip('\n ')
    T.append(h)
conceded=soup.findAll('div',attrs={'class':'sipk-lb-detailBlock sipk-lb-raidPoints'})
P=[]
for i in conceded:
    k=i.text
    k=k.strip('\n')
    k=k.strip(' ')
    k=k.strip('\n ')
    P.append(k)
    
Season7_conceded=pd.DataFrame(T,P)
Season7_conceded=Season7_conceded.reset_index()
Season7_conceded=Season7_conceded.rename(columns={'index':'Conceded',0        :'Team'})
Season7_conceded=Season7_conceded[['Team','Conceded']]
```

```python
from functools import reduce

Season7=[Season7_TeamPoints,Season7_TeamPointsConceded,Season7_successful_raids,Season7_successful_tackles,Season7_super_raid,Season7_super_tackles,Season7_inflicted,Season7_conceded]

Season7=reduce(lambda x,y:pd.merge(x,y,on='Team'),Season7)
```
